```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.llms import Ollama
from langchain_core.output_parsers import StrOutputParser

# Set up the prompt template for JSON formatting
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in the field. With great analytical skills with focus on a clear and structured communication.You always use data available to back up all your claims. At the start of your answer ALWAYS Provide a yes/no prediction, a probability score, and a confidence level in a JSON format. And then continue with text on the explanation of why you said what you said."),
    ("user", "{input}")
])

```

```python
# Initialize mistral with Ollama
llm = Ollama(model="mistral")
# Combine prompt template with LLM and output parser
chain = prompt | llm | StrOutputParser()
```

```python
# Initialize llama2 with Ollama
llm = Ollama(model="llama2")
```

```python
question_answers = []

def invoke_llm(question):
    # Invoke the chain with the question
    response = chain.invoke({"input": question})
    question_answers.append((question, response))
```

```python
# Example usage of the invoke_llm function
invoke_llm("Given the current economic trends and market data up to January 2024, what is the probability of the Federal Reserve increasing interest rates by the end of Q2 2024?")
invoke_llm("Considering the latest advancements and regulatory status in autonomous vehicle technology, what is the likelihood of a major TESLA launching a fully autonomous vehicle for consumer use by 2026?")
invoke_llm("Based on recent advancements in renewable energy technology and current investment trends, what is the probability that solar energy will account for 30% of global energy production by 2030?")
invoke_llm("Analyze the current trends in the global smartphone market. What is the likelihood of Xiaomi becoming the market leader in terms of sales volume by the end of the next fiscal year?")
invoke_llm("Considering the current geopolitical landscape and diplomatic efforts, will there be a significant reduction in global carbon emissions by 2030 in line with international climate agreements? ")
invoke_llm("Analyze the trends in the cryptocurrency market. Do you predict that Bitcoin will reach a price of $100,000 per coin within the next year?")
```

```python
# Print all questions and their corresponding answers
for question, response in question_answers:
    print(f"Question: {question}")
    print(f"Answer:\n{response}")
    print("\n")
```

```python
# Measuring accuracy of responds in historical data
invoke_llm("Predict the outcome of the 2016 United States presidential election. What was the probability of Donald Trump winning, and provide a confidence level for your prediction.")
invoke_llm("What was the likelihood of the Los Angeles Lakers winning the 2020 NBA Championship? Provide a probability score and a confidence level for your prediction.")
invoke_llm("As of the end of 2019, what was the probability of the US entering a recession in 2020? Provide a binary prediction, probability score, and confidence level.")
invoke_llm("As of the beginning of 2020, what was the likelihood of a COVID-19 vaccine being available for public use by the end of the year? Provide a binary prediction, probability score, and confidence level.")
```
